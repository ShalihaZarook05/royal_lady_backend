```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
df = pd.read_csv("myntra_products_catalog.csv")
df.head()
```

```python
df["combined"] = (
    df["ProductName"].fillna("") + " " +
    df["ProductBrand"].fillna("") + " " +
    df["Description"].fillna("") + " " +
    df["PrimaryColor"].fillna("") + " " +
    df["Gender"].fillna("")
)
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
product_vectors = vectorizer.fit_transform(df["combined"])
```

```python
similarity_matrix = cosine_similarity(product_vectors)
```

```python
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))
pickle.dump(similarity_matrix, open("similarity.pkl", "wb"))
pickle.dump(df, open("product_data.pkl", "wb"))

print("AI Model Training Completed Successfully!")
```
