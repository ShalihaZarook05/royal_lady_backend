```python
import pandas as pd
import os

print("Current directory:", os.getcwd())

df = pd.read_csv("myntra_products_catalog.csv")

df = df.dropna(subset=["ProductName", "Description"])
df = df.reset_index(drop=True)

df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
product_vectors = vectorizer.fit_transform(df["Description"])
```

```python
import pickle

with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

with open("product_vectors.pkl", "wb") as f:
    pickle.dump(product_vectors, f)

with open("product_data.pkl", "wb") as f:
    pickle.dump(df.to_dict(orient="records"), f)

print("✅ AI model regenerated successfully")
```
